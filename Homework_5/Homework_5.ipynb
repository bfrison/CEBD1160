```python
#!/usr/bin/env python3

%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D

import itertools
import os
os.chdir('/home/bfrison/Homework_5/')
```

```python
df = pd.read_csv('diabetes.data', sep='\s+', header=0)
fig, axes = plt.subplots(15, 3, figsize=(27, 135))
k = 0

df_1 = df[df['SEX'] == 1]
df_2 = df[df['SEX'] == 2]
df_no_sex = df[(col for col in df.columns if col != 'SEX')]

for i, col1 in enumerate(df_no_sex):
    for j, col2 in enumerate(df_no_sex):
        if col1 < col2:
            ax = axes[int(k/3)][k%3]
            k += 1
            sns.scatterplot(df_1[col1], df_1[col2], ax=ax, label='1')
            sns.scatterplot(df_2[col1], df_2[col2], ax=ax, label='2')
            ax.set_title(f'Scatter plot of {col1} vs {col2} separated by sex')
            ax.legend()
```
