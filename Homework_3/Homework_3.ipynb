```python
# 3.
#!/usr/bin/env python3
import random
debug = False
tries = 0
LIMIT = 5
r = random.Random()
computer_number = r.randint(0, 200)
user_number = None

def print_clue(number):
    if number % 2 == 0:
        print('Number is even')
    else:
        print('Number is odd')
    if number % 3 == 0:
        print('Number is divisible by 3\n')
    else:
        print('Number is not divisible by 3\n')

while user_number != computer_number:
    tries += 1
    if debug:
        print('Computer choice is {:d}'.format(computer_number))
    user_number = input('Please type a number between 0 and 200:\n')
    try:
        user_number = int(user_number)
    except ValueError:
        user_number = None
    if user_number == computer_number:
        print('\nBingo!')
    elif tries < LIMIT:
        if user_number is None:
            print('\nInvalid number.\n')
        elif computer_number - user_number >= 50:
            print('\nWay upper\n')
        elif computer_number - user_number >= 2:
            print('\nUpper')
            if tries > 1:
                print_clue(computer_number)
            else:
                print('')
        elif user_number - computer_number >= 50:
            print('\nWay lower\n')
        elif user_number - computer_number >= 2:
            print('\nLower')
            if tries > 1:
                print_clue(computer_number)
            else:
                print('')
        elif abs(user_number - computer_number) == 1:
            print('\nClose!\n')
    else:
        print('\nNo more tries!')
        break
```
